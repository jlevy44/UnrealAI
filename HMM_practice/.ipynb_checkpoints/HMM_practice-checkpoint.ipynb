```python
%matplotlib inline 
import matplotlib

import pandas as pd
import networkx as nx
import numpy as np
from collections import Counter
import functools
import matplotlib.pyplot as plt
from networkx.drawing.nx_agraph import write_dot
```

```python
E = pd.DataFrame(columns=list('DNAVP'),index='the some wine flies dove last ages waits alone into for'.split()).fillna(0)
```

```python
state_emission_pairs = list(filter(lambda x: x!=tuple(['']),map(lambda x: tuple(x.strip('( \n').split(',')),"""(D,the) (N,wine) (V,ages) (A,alone)
(D,the) (N,wine) (N,waits) (V,last) (N,ages)
(D,some) (N,flies) (V,dove) (P,into) (D,the) (N,wine)
(D,the) (N,dove) (V,flies) (P,for) (D,some) (N,flies)
(D,the) (A,last) (N,dove) (V,waits) (A,alone)""".split(')'))))
```

```python
for state, emission in state_emission_pairs:
    E[state][emission] += 1
```

```python
E/=E.sum(0)
print(E)
```

```python
state_transitions = functools.reduce(lambda a,b: a+b,map(lambda y: [(y[i],y[i+1]) for i in range(len(y)-1)], ['DNNVN','DNVPDN','DNVPDN','DANVA']))
```

```python
T = pd.DataFrame(columns=list('DNAVP'),index=list('DNAVP')).fillna(0)
for s1,s2 in state_transitions:
    T[s1][s2] += 1
T /= T.sum(axis=0)
```

```python
T
```

```python
print(T.to_dict())
G = nx.DiGraph(T.to_dict())
G2 = nx.DiGraph(E.to_dict())
G = nx.compose(G,G2)
```

```python
plt.figure()
nx.draw(G)
write_dot(G,'graph.dot')
```

```python
States = {}

for state in list(E):
    #print(E[state].to_dict())
    States[state] = State(DiscreteDistribution(E[state].to_dict()),name=state)
    
hmm = HiddenMarkovModel()
hmm.add_states(*list(States.values()))

for state in list(T):
    hmm.add_transition(hmm.start, States[state], T[state][state])
    hmm.add_transition(States[state],States[state], T[state][state])
    
#print(hmm.states)
#print(States)
    
for s1,s2 in permutations(list(T),r=2):
    hmm.add_transition(States[s1],States[s2], T[s1][s2])

#hmm.edges=T.as_matrix().tolist()

hmm.bake()
    
```

```python
seq = 'dove waits for the wine'.split()# https://github.com/jmschrei/pomegranate/blob/master/tutorials/Tutorial_3_Hidden_Markov_Models.ipynb
POS = list(E)#.unique()
#v_logp, v_seq = hmm.viterbi(seq)
#v_logp, v_seq
#hmm.fit(sequences,baum_welch) can adjust parameters
hmm_predictions = hmm.predict(seq , algorithm='viterbi')[1:]
print("sequence: {}".format( ' '.join( seq ) ))
print("hmm pred: {}".format( ' '.join( map( lambda i: POS[i-1], hmm_predictions ) ) ))# POS[i] str(i)
```

```python
seq = list('CGACTACTGACTACTCGCCGACGCGACTGCCGTCTATACTGCGCATACGGC') # FIXME change MLE or EM estimation of T and E, because training sequences are sampled from a distribution

d1 = DiscreteDistribution({'A': 0.25, 'C': 0.25, 'G': 0.25, 'T': 0.25})
d2 = DiscreteDistribution({'A': 0.10, 'C': 0.40, 'G': 0.40, 'T': 0.10})

s1 = State( d1, name='background' )
s2 = State( d2, name='CG island' )

hmm = HiddenMarkovModel()
hmm.add_states(s1, s2)
hmm.add_transition( hmm.start, s1, 0.5 )
hmm.add_transition( hmm.start, s2, 0.5 )
hmm.add_transition( s1, s1, 0.5 )
hmm.add_transition( s1, s2, 0.5 )
hmm.add_transition( s2, s1, 0.5 )
hmm.add_transition( s2, s2, 0.5 )
hmm.bake()

hmm_predictions = hmm.predict( seq )

print("sequence: {}".format( ''.join( seq ) ))
print("hmm pred: {}".format( ''.join( map( str, hmm_predictions ) ) ))
```

```python
trans, ems = hmm.forward_backward( seq )
```
